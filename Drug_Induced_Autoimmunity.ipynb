```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split,RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
import matplotlib.pyplot as plt
```

```python
train_file =r"C:\Users\anush_jg2uovv\OneDrive\Documents\4th SEM\ML\Assignment 1\DIA_trainingset_RDKit_descriptors.csv"
test_file =r"C:\Users\anush_jg2uovv\OneDrive\Documents\4th SEM\ML\Assignment 1\DIA_testset_RDKit_descriptors.csv"

df_train = pd.read_csv(train_file)
df_test = pd.read_csv(test_file)
```

```python
x_train=df_train.drop(columns=['Label','SMILES'])
y_train=df_train['Label']

x_test=df_test.drop(columns=['Label','SMILES'])
y_test=df_test['Label']
```

```python
x_train.head()
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
#Model1- Decision tree
from sklearn.tree import DecisionTreeClassifier
param_dist = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30, 40],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None,'sqrt', 'log2']
}

dt = DecisionTreeClassifier(random_state=42)
random_search = RandomizedSearchCV(estimator=dt, param_distributions=param_dist, n_iter=20, cv=5, random_state=42, n_jobs=-1)
random_search.fit(x_train_scaled, y_train)
y_pred = random_search.best_estimator_.predict(x_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
print("\nClassification Report:", class_report)
print("\nConfurion Matrix:", conf_matrix)
```

```python
#Model2- Support Vector Classifier
from sklearn.svm import SVC
param_dist = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['rbf', 'linear', 'poly', 'sigmoid']
}

svc = SVC(random_state=42)
random_search = RandomizedSearchCV(estimator=svc, param_distributions=param_dist, n_iter=20, cv=5, random_state=42, n_jobs=-1, verbose=2)
random_search.fit(x_train_scaled, y_train)
y_pred = random_search.best_estimator_.predict(x_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
print("\nClassification Report:", class_report)
print("\nConfurion Matrix:", conf_matrix)
```

```python
#Model3- Logistic Regression
from sklearn.linear_model import LogisticRegression
param_dist = [
    {'C': [0.01, 0.1, 1, 10, 100], 'penalty': ['l1', 'l2'], 'solver': ['liblinear']},
    {'C': [0.01, 0.1, 1, 10, 100], 'penalty': ['elasticnet'], 'solver': ['saga'], 'l1_ratio': [0.0, 0.25, 0.5, 0.75, 1.0]}
]

logreg = LogisticRegression(max_iter=1000, random_state=42)
random_search = RandomizedSearchCV(logreg, param_distributions=param_dist, n_iter=20, cv=5, scoring='accuracy', n_jobs=-1,
    random_state=42, error_score='raise', refit=True)
random_search.fit(x_train_scaled, y_train)
y_pred = random_search.best_estimator_.predict(x_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
print("\nClassification Report:", class_report)
print("\nConfurion Matrix:", conf_matrix)
```

```python
#Model4- Random forest
from sklearn.ensemble import RandomForestClassifier
param_dist = {
    'n_estimators': np.arange(50, 201, 50),
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'bootstrap': [True, False]
}

rf = RandomForestClassifier(random_state=42)
random_search = RandomizedSearchCV(estimator=rf, param_distributions=param_dist, n_iter=20, cv=5, random_state=42, n_jobs=-1)
random_search.fit(x_train_scaled, y_train)
y_pred = random_search.best_estimator_.predict(x_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
print("\nClassification Report:", class_report)
print("\nConfurion Matrix:", conf_matrix)
```
